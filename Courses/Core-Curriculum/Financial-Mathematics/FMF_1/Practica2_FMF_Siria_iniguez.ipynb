**Parte I: Contrato Forward**

**Parte II: Contrato Futuro con Marking-to-Market**

Considera una curva de tipos plana al  8%  con composici√≥n continua. Calcula los precios futuros de un activo cuyo precio actual es  ùëÜ0=100 , y los valores del ‚Äúmarking to market‚Äù para un contrato futuro con vencimiento en dos a√±os y ajustes semestrales, asumiendo que los precios del activo son

```python
Si = [100, 110, 105, 94, 97]  # Precios spot en T0, T1, T2, T3, T4
r = 0.08  # 8% continua
T = 2.0   # Vencimiento en 2 a√±os (NO 1 como ten√≠as)

# Fechas: semestrales, empezando en 0
fechas = [0.0, 0.5, 1.0, 1.5, 2.0]  # T_i = i/2 a√±os

# Calcular precios futuros
Ft = []
for i in range(len(Si)):
    t = fechas[i]  # Tiempo actual
    tiempo_hasta_vencimiento = T - t  # T - t
    F = Si[i] * math.exp(r * tiempo_hasta_vencimiento)
    Ft.append(F)

print("Precios futuros F(t,2):")
print([round(f, 6) for f in Ft])

MTM_i = []
for i in range (1, len(Si)):
  MTM_i.append(Ft[i]-Ft[i-1])
print("\n Marking to market: ")
print(MTM_i)
```
